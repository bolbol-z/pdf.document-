# 🍃 MongoDB — Référentiel Compétences

**Target:** MongoDB 6.0+  |  **Encoding:** UTF-8  |  **Connection:** `mongodb://localhost:27017`

| Collection | Strategy | Documents |
|---|---|---|
| `metiers` | Embed competences | 24 |
| `domaines` | Embed metier refs | 7 |
| `vocabulaire` | Unified lookups | 542 |

**Design:** SQL 12 tables + 6 junctions → 3 MongoDB collections via embedding + selective referencing.
- Competences are **owned** by metier → embedded (single `findOne` per query)
- Outils / mots_cles / normes → embedded string arrays (`$in`, `$elemMatch`)
- Vocabulary lookups → separate collection, filtered by `categorie` field

## 0. Setup — Connect & Select Database

```python
from pymongo import MongoClient, ASCENDING, TEXT
import json, pprint

# ── Connection ────────────────────────────────────────────────────────────────
MONGO_URI = "mongodb://localhost:27017"
DB_NAME   = "referentiel_competences"

client = MongoClient(MONGO_URI)
db     = client[DB_NAME]

print(f"✅ Connected  : {MONGO_URI}")
print(f"   Database   : {DB_NAME}")
print(f"   Existing collections: {db.list_collection_names()}")
```

## 1. Create Collections with Schema Validation

```python
# Drop and recreate for a clean run
for col in ['metiers', 'domaines', 'vocabulaire']:
    db.drop_collection(col)
    print(f"  Dropped: {col}")

# ── metiers ───────────────────────────────────────────────────────────────────
db.create_collection('metiers', validator={
    '$jsonSchema': {
        'bsonType': 'object',
        'required': ['_id', 'metier_id', 'nom_metier', 'domaine', 'competences'],
        'properties': {
            'metier_id':      {'bsonType': 'int',    'description': 'SQL FK reference'},
            'nom_metier':     {'bsonType': 'string', 'description': 'Unique metier name'},
            'domaine':        {'bsonType': 'object', 'description': 'Embedded domaine snapshot'},
            'nb_competences': {'bsonType': 'int',    'description': 'Denormalized count'},
            'competences':    {'bsonType': 'array',  'description': 'Embedded competence documents'},
        }
    }
})

# ── domaines ──────────────────────────────────────────────────────────────────
db.create_collection('domaines', validator={
    '$jsonSchema': {
        'bsonType': 'object',
        'required': ['_id', 'domaine_id', 'nom_domaine', 'metiers'],
        'properties': {
            'domaine_id':  {'bsonType': 'int'},
            'nom_domaine': {'bsonType': 'string'},
            'nb_metiers':  {'bsonType': 'int'},
            'metiers':     {'bsonType': 'array'},
        }
    }
})

# ── vocabulaire ───────────────────────────────────────────────────────────────
db.create_collection('vocabulaire', validator={
    '$jsonSchema': {
        'bsonType': 'object',
        'required': ['_id', 'categorie', 'libelle'],
        'properties': {
            'categorie': {
                'bsonType': 'string',
                'enum': ['type_competence', 'modalite_evaluation', 'formation_activite',
                         'outil', 'reglementation_norme', 'mot_cle']
            },
            'libelle': {'bsonType': 'string'},
            'code_id': {'bsonType': 'int'},
        }
    }
})

print("\n✅ Collections created:")
for col in db.list_collection_names():
    print(f"   {col}")
```

## 2. Indexes

```python
# ── metiers ───────────────────────────────────────────────────────────────────
db.metiers.create_index([('domaine.domaine_id', ASCENDING)],       name='idx_metier_domaine')
db.metiers.create_index([('competences.type_competence', ASCENDING)], name='idx_metier_comp_type')
db.metiers.create_index([('competences.mots_cles', ASCENDING)],    name='idx_metier_mots_cles')
db.metiers.create_index([('competences.outils', ASCENDING)],       name='idx_metier_outils')
db.metiers.create_index(
    [('nom_metier', TEXT), ('competences.libelle', TEXT),
     ('competences.mots_cles', TEXT), ('competences.indicateurs_observables', TEXT)],
    name='idx_metier_fulltext',
    weights={'nom_metier': 10, 'competences.libelle': 8,
             'competences.mots_cles': 5, 'competences.indicateurs_observables': 2},
    default_language='french'
)

# ── domaines ──────────────────────────────────────────────────────────────────
db.domaines.create_index([('domaine_id', ASCENDING)], name='idx_domaine_id', unique=True)

# ── vocabulaire ───────────────────────────────────────────────────────────────
db.vocabulaire.create_index([('categorie', ASCENDING), ('libelle', ASCENDING)],
                             name='idx_vocab_cat_lib')
db.vocabulaire.create_index([('libelle', TEXT)],
                             name='idx_vocab_fulltext', default_language='french')

print("✅ Indexes created:")
for col in ['metiers', 'domaines', 'vocabulaire']:
    for idx in db[col].list_indexes():
        print(f"   [{col}] {idx['name']}")
```

## 6. Example Queries

### Q1 — Full profile of a metier

```python
doc = db.metiers.find_one({'nom_metier': 'Conducteur Routier (PL/SPL)'})
# Show header + first 2 competences
preview = {k: v for k, v in doc.items() if k != 'competences'}
preview['competences'] = doc['competences'][:2]
print(json.dumps(preview, ensure_ascii=False, indent=2))
```

### Q2 — All competences for a metier

```python
doc = db.metiers.find_one(
    {'nom_metier': 'Conducteur Routier (PL/SPL)'},
    {'nom_metier': 1, 'competences': 1, '_id': 0}
)
print(f"Metier: {doc['nom_metier']}  |  {len(doc['competences'])} compétences")
for c in doc['competences']:
    print(f"  • [{c['type_competence']}] {c['libelle'][:70]}...")
```

### Q3 — Filter competences by type within a metier

```python
pipeline = [
    {'$match': {'nom_metier': 'Mécanicien Poids Lourds'}},
    {'$unwind': '$competences'},
    {'$match': {'competences.type_competence': 'Technique'}},
    {'$replaceRoot': {'newRoot': '$competences'}}
]
results = list(db.metiers.aggregate(pipeline))
print(f"Compétences Technique — Mécanicien Poids Lourds: {len(results)}")
for r in results:
    print(f"  • {r['libelle'][:80]}")
```

### Q4 — Search metiers by mots_cles

```python
keywords = ['sécurité', 'coordination']
results = list(db.metiers.find(
    {'competences.mots_cles': {'$in': keywords}},
    {'nom_metier': 1, 'domaine.nom_domaine': 1, 'nb_competences': 1, '_id': 0}
))
print(f"Metiers with keywords {keywords}: {len(results)}")
for r in results:
    print(f"  • {r['nom_metier']}  [{r['domaine']['nom_domaine']}]")
```

### Q5 — Find all metiers using a specific tool

```python
tool = 'WMS'
results = list(db.metiers.find(
    {'competences.outils': tool},
    {'nom_metier': 1, 'domaine.nom_domaine': 1, '_id': 0}
))
print(f"Metiers using '{tool}': {len(results)}")
for r in results:
    print(f"  • {r['nom_metier']}")
```

### Q6 — Full-text search across competences

```python
results = list(db.metiers.find(
    {'$text': {'$search': 'conduite sécurité', '$language': 'french'}},
    {'score': {'$meta': 'textScore'}, 'nom_metier': 1, '_id': 0}
).sort([('score', {'$meta': 'textScore'})]).limit(5))

print("Full-text search: 'conduite sécurité'")
for r in results:
    print(f"  • {r['nom_metier']}  (score: {r['score']:.2f})")
```

### Q7 — Matching system: score metiers by CV keyword overlap

```python
cv_skills = ['coordination', 'stress', 'autonomie', 'Excel', 'WMS']

pipeline = [
    {'$unwind': '$competences'},
    {'$unwind': '$competences.mots_cles'},
    {'$match': {'competences.mots_cles': {'$in': cv_skills}}},
    {'$group': {
        '_id': '$_id',
        'nom_metier':       {'$first': '$nom_metier'},
        'domaine':          {'$first': '$domaine.nom_domaine'},
        'matches':          {'$sum': 1},
        'matched_keywords': {'$addToSet': '$competences.mots_cles'}
    }},
    {'$sort': {'matches': -1}},
    {'$limit': 5}
]

results = list(db.metiers.aggregate(pipeline))
print(f"CV skills: {cv_skills}")
print(f"Top {len(results)} matching metiers:\n")
for r in results:
    print(f"  🏆 {r['nom_metier']}")
    print(f"     Domain : {r['domaine']}")
    print(f"     Score  : {r['matches']} matches")
    print(f"     Keywords matched: {r['matched_keywords']}\n")
```

### Q8 — Metiers per domaine with competence count

```python
pipeline = [
    {'$group': {
        '_id': '$domaine.nom_domaine',
        'metiers': {'$push': {'nom': '$nom_metier', 'nb_comp': '$nb_competences'}},
        'total_competences': {'$sum': '$nb_competences'}
    }},
    {'$sort': {'_id': 1}}
]
for r in db.metiers.aggregate(pipeline):
    print(f"\n📁 {r['_id']}  (total compétences: {r['total_competences']})")
    for m in r['metiers']:
        print(f"   • {m['nom']}  ({m['nb_comp']} comp.)")
```

### Q9 — Top 10 most common mots_cles

```python
pipeline = [
    {'$unwind': '$competences'},
    {'$unwind': '$competences.mots_cles'},
    {'$group': {'_id': '$competences.mots_cles', 'count': {'$sum': 1}}},
    {'$sort': {'count': -1}},
    {'$limit': 10}
]
print("Top 10 mots-clés:")
for i, r in enumerate(db.metiers.aggregate(pipeline), 1):
    print(f"  {i:2}. {r['_id']:<25} ({r['count']}×)")
```
